```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
a, b = -2, 2
M = 100
h = (b-a)/M
x = np.arange(a, b+h, h)
y = np.exp(x)
plt.plot(x, y);
```

We're trying to approximate

$$\int_{-2}^2 e^x dx = e^2 - e^{-2}$$

```python
exact = np.exp(2) - np.exp(-2)
```

```python
approx = integrate.trapezoid(y, dx=h)
abs(exact - approx)
```

The cumulative trapezoidal approximates the indefinite integral $\int e^x dx$ evaluated at grid points $\{x_i\}$

```python
cum_approx = integrate.cumulative_trapezoid(y, dx=h, initial=0)
plt.plot(x, cum_approx)
plt.plot(x, y);
```

```python
cum_approx[-1] - approx
```

The last point of the cumulative array is the same as the result of the non-cumulative version, as expected.
